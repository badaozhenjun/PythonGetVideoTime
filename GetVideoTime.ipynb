```python
import os
import sys
from moviepy.editor import VideoFileClip
import xlsxwriter
# 获取文件大小
def get_file_size(filename):
    file_byte = os.path.getsize(filename)
    return sizeConvert(file_byte)

# 获取视频时长
def get_file_times(filename):
    clip = VideoFileClip(filename)
    file_time = timeConvert(clip.duration)
    return file_time

# 文件大小单位换算 Bytes-->K,M,G
def sizeConvert(size):
    K, M, G = 1024, 1024 ** 2, 1024 ** 3
    if size >= G:
        return str(size / G) + 'G Bytes'
    elif size >= M:
        return str(size / M) + 'M Bytes'
    elif size >= K:
        return str(size / K) + 'K Bytes'
    else:
        return str(size) + 'Bytes'

# 时间转换 seconds-->小时分钟秒
def timeConvert(size):
    M, H = 60, 60 ** 2
    return int(size / M)
    if size < M:
        return str(size) + u'秒'
    if size < H:
        return u'%s分钟%s秒' % (int(size / M), int(size % M))
    else:
        hour = int(size / H)
        mine = int(size % H / M)
        second = int(size % H % M)
        tim_srt = u'%s小时%s分钟%s秒' % (hour, mine, second)
        return tim_srt

# 获取当前路径下所有非目录子文件
def get_files_in_path(file_path):
    for root, dirs, files in os.walk(file_path):
        return files

#获取当前路径下以xxx结尾的所有文件
def get_all_files_by_ends(file_path,ends):
    videos = []
    g = os.walk(file_path)
    for path, d, filelist in g:
        for filename in filelist:
            if filename.endswith(ends):
                videos.append(os.path.join(path, filename))

    return videos

# 获取当前目录下所有视频文件的时长,并返回字典
def get_videos_length(file_path):
    videos = get_all_files_by_ends(file_path, ".wmv")
    videos_dic={}
    i = 0
    length = len(videos)
    #可能由于内存溢出导致程序崩溃
    try:
        while i<length:
            videos_dic[videos[i]] = get_file_times(videos[i])
            i +=1
    except:
        #打印异常信息
        print(sys.exc_info()[0],sys.exc_info()[1])
    finally:
        return videos_dic

# 测试
if __name__ == ("__main__"):
    # 视频目录在此修改
    file_path = "/Users/yuxiao/Downloads/VIP1/《Shark恒零基础百集VIP破解教程》 - 高清无密官方原版"
    workbook = xlsxwriter.Workbook(file_path+'/time.xlsx')
    worksheet = workbook.add_worksheet()
    
    videos_dic=get_videos_length(file_path)
    i=0
    for key in videos_dic:
        worksheet.write(i,1,videos_dic[key])
        names = key.split("/")
        filename=names[len(names)-1]
        worksheet.write(i,0,filename)
        i=i+1
        print(str(videos_dic[key])+str(":")+filename)
        #print(videos_dic[key]) 
    worksheet.write(i,0,"合计")
    worksheet.write(i,1,str('%.2f'%(sum(list(videos_dic.values()))/60)+"小时"))
    workbook.close()
```
